```python
df['movie'].str.strip()
```

```python
#REMOVE ANY PLACEHOLDERS IN THE INTEGER COLUMNS
df['domestic_gross'] = df['domestic_gross'].replace('[\$,]', '', regex=True)
df['worldwide_gross'] = df['worldwide_gross'].replace('[\$,]', '', regex=True)
df['popularity'] = df['popularity'].replace('[\$,]', '', regex=True).astype(float)
df['vote_average'] = df['vote_average'].replace('[\$,]', '', regex=True).astype(float)
df['vote_count'] = df['vote_count'].replace('[\$,]', '', regex=True).astype(float)
```

```python
def remove_outliers(df, column='columns'):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df

```
